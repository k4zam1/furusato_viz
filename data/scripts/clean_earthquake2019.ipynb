```python
import pandas as pd
df = pd.read_csv("earthquake_2019.csv")
```

```python
df.head()
```

```python
import re
def DMStoDMG(dms):
    """度分秒表記の緯度経度を10進数表記に変換する"""
    splits = re.split('[°\'′".]+', dms)
    print(splits)
    if len(splits) == 4:
        degrees,minutes,seconds,_ = splits
    elif len(splits) == 3:
        degrees,minutes,_ = splits
        seconds = 0
    else :
        return 0
    dmg = float(degrees) + float(minutes)/60 + float(seconds)/(60*60);
    return dmg
```

```python
df["緯度"] = df["緯度"].map(DMStoDMG)
df["経度"] = df["経度"].map(DMStoDMG)
df.head()
```

```python
df.to_csv("cleaned_earthquake2019.csv",index=False)
```
